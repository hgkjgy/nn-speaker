# Converting a trained model to tflite
https://www.tensorflow.org/lite/microcontrollers/build_convert#model_conversion

# Convert model to tflite

```python
from pathlib import Path

input_file = Path("converted_model.tflite")
output_file = Path("model_data.cc")

if not input_file.exists():
    raise FileNotFoundError(f"Cannot find {input_file}")

data = input_file.read_bytes()

array_name = "converted_model_tflite"
line_width = 12

with output_file.open("w", encoding="utf-8") as f:
    f.write("#include <cstddef>\n")
    f.write("#include <cstdint>\n\n")
    f.write(f"unsigned char {array_name}[] = {{\n")

    for i in range(0, len(data), line_width):
        chunk = data[i:i + line_width]
        hex_bytes = ", ".join(f"0x{b:02x}" for b in chunk)
        f.write(f"    {hex_bytes},\n")

    f.write("};\n")
    f.write(f"unsigned int {array_name}_len = {len(data)};\n")
```

# To convert to C++
This will run a command line too to convert out tflite model into C code.

```python
from pathlib import Path

input_file = Path("converted_model.tflite")
output_file = Path("model_data.cc")

if not input_file.exists():
    raise FileNotFoundError(f"Cannot find {input_file}")

data = input_file.read_bytes()

array_name = "g_model_data"
line_width = 12

with output_file.open("w", encoding="utf-8") as f:
    f.write("#include <cstddef>\n")
    f.write("#include <cstdint>\n\n")
    f.write(f"const unsigned char {array_name}[] = {{\n")

    for i in range(0, len(data), line_width):
        chunk = data[i:i + line_width]
        hex_bytes = ", ".join(f"0x{b:02x}" for b in chunk)
        f.write(f"    {hex_bytes},\n")

    f.write("};\n")
    f.write(f"const unsigned int {array_name}_len = {len(data)};\n")
```
